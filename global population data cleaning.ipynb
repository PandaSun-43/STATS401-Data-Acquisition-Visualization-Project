```python
import pandas as pd
```

```python
df = pd.read_csv('population.csv')
df.head()
```

```python
df = df[df['Entity'].isin(['Asia', 'Europe', 'Africa', 'Oceania', 'North America', 'South America'])]
df = df[(df['Year'] <= 2023) & (df['Year'] >= 1920)]
global_population = df.groupby("Year")["Population (historical)"].sum()

# Merge the computed global population back into the dataframe
df["Global Annual Population (billions)"] = df["Year"].map(global_population)
df = df[df['Entity'] == 'Asia']
df = df.drop(columns=['Entity', 'Population (historical)', 'Code']).reset_index(drop=True)
df["Global Annual Population (billions)"] /= 1e9
df.head()
```

```python
df.to_csv('cleaned_global_annual_population.csv', index=False)
```
