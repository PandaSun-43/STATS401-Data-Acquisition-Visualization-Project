```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load your data into a DataFrame
df = pd.read_csv("cleaned_annual-co-emissions-country.csv")

# Prepare the time series data
df = df[['Entity', 'Year', 'Annual CO_ emissions']]

# Initialize an empty dictionary to store forecasts
forecast_dict = {}

# Loop through each unique country (Entity)
for country in df['Entity'].unique():
    country_data = df[df['Entity'] == country]
    
    # Set Year as the independent variable (X) and 'Change in global mean surface temperature' as the dependent variable (y)
    X = country_data[['Year']]  # We only use the Year to predict the temperature change
    y = country_data['Annual CO_ emissions']
    
    # Instantiate and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Make predictions for the next 7 years (2024-2030)
    future_years = np.array(range(2024, 2031)).reshape(-1, 1)
    forecast = model.predict(future_years)
    
    # Store the forecast in the dictionary
    forecast_dict[country] = forecast[6]
    

# Print the forecast
forecast_df = pd.DataFrame(list(forecast_dict.items()), columns=['Country', 'Forecast'])
print(forecast_dict)
```

```python
forecast_df.to_csv("forecast_cm.csv",index = False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load your data into a DataFrame
df = pd.read_csv("cleaned_population.csv")
df = df[df['Entity'].isin(name_list)]
df = df.drop(columns = 'Code')

# Prepare the time series data
df = df[['Entity', 'Year', 'Population (historical)']]

# Initialize an empty dictionary to store forecasts
forecast_dict = {}

# Loop through each unique country (Entity)
for country in df['Entity'].unique():
    country_data = df[df['Entity'] == country]
    
    # Set Year as the independent variable (X) and 'Change in global mean surface temperature' as the dependent variable (y)
    X = country_data[['Year']]  # We only use the Year to predict the temperature change
    y = country_data['Population (historical)']
    
    # Instantiate and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Make predictions for the next 7 years (2024-2030)
    future_years = np.array(range(2024, 2031)).reshape(-1, 1)
    forecast = model.predict(future_years)
    
    # Store the forecast in the dictionary
    forecast_dict[country] = forecast[6]
    

# Print the forecast
forecast_df = pd.DataFrame(list(forecast_dict.items()), columns=['Country', 'Forecast'])
print(forecast_dict)
```
