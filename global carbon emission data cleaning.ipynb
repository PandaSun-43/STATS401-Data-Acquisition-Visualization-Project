```python
import pandas as pd
```

```python
df = pd.read_csv('annual-co-emissions.csv')
df.head()
```

```python
df = df[df['Entity'].isin(['Asia', 'Europe', 'Africa', 'Oceania', 'North America', 'South America'])]
df = df[(df['Year'] <= 2023) & (df['Year'] >= 1920)]
global_co2 = df.groupby("Year")["Annual CO₂ emissions"].sum()

df["Global Annual CO2 Emissions (billion t)"] = df["Year"].map(global_co2)
df = df[df['Entity'] == 'Asia']
df = df.drop(columns=['Entity', 'Annual CO₂ emissions']).reset_index(drop=True)
df['Global Annual CO2 Emissions (billion t)'] /= 1e9
df.head()
```

```python
df.to_csv('cleaned_global_annual_co2_emission.csv', index=False)
```
