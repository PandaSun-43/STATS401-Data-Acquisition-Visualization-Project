```python
import pandas as pd
```

```python
df = pd.read_csv('temperature-anomaly.csv')
df.head()
```

```python
df = df[df['Entity'] == 'Global']
```

```python
df.drop(columns=['Code', 'Entity', 'Upper bound of the annual temperature anomaly (95% confidence interval)', 'Lower bound of the annual temperature anomaly (95% confidence interval)'], inplace=True)
df = df[df['Year'] >= 1920]
df = df[df['Year'] <= 2023]
```

```python
df = df.reset_index(drop=True)
df.to_csv('cleaned_temperature_anomaly.csv', index=False)
```
