```python
import pandas as pd

# Load the dataset
df = pd.read_csv("cleaned_annual-co-emissions-country.csv")

# Display column names to check for inconsistencies
print(df.columns)
```

```python
# Rename columns if necessary (fix spacing, encoding issues)
df.rename(columns=lambda x: x.strip(), inplace=True)

# Convert "Annual CO₂ emissions" to numeric (handle missing or incorrect values)
df["Annual CO2 emissions"] = pd.to_numeric(df["Annual CO2 emissions"], errors="coerce")
df
```

```python
df = df[df['Entity'] != 'World']
df
```

```python

df.to_csv("cleaned_annual-co-emissions-country.csv", index=False)
```
